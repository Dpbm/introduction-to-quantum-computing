```python
%matplotlib inline
from qiskit import QuantumCircuit, transpile
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_histogram, plot_bloch_multivector
from qiskit_aer import AerSimulator
import matplotlib.pyplot as plt
```

```python
def get_filename(name):
    return "images/"+name
```

```python
def show_and_save_bloch(qc, filename):
    bloch = plot_bloch_multivector(qc)
    bloch.savefig(get_filename(filename))
    display(bloch)
```

```python
backend = AerSimulator()
```

---

# Pauli X Gate

```python
x_gate = QuantumCircuit(1)
x_gate.draw('mpl', filename=get_filename("empty_circuit.png"))
```

```python
show_and_save_bloch(x_gate, "empty_circuit_bloch_sphere.png")
```

```python
x_gate.x(0)
x_gate.draw('mpl', filename=get_filename("x_gate.png"))
```

```python
show_and_save_bloch(x_gate, "x_gate_bloch_sphere.png")
```

---

# Hadamard Gate

```python
h_gate = QuantumCircuit(1)
h_gate.h(0)
h_gate.draw('mpl', filename=get_filename("h_gate.png"))
```

```python
show_and_save_bloch(h_gate, "h_gate_bloch_sphere.png")
```

```python
def show_and_save_histogram(results, filename):
    plot = plot_histogram(results)
    plot.savefig(get_filename(filename))
    display(plot)
```

```python
def execute(qc, sim=backend, shots=1000):
    results = sim.run(transpile(qc, sim), shots=shots).result().get_counts()
    return results
```

```python
h_gate.measure_all()
results = execute(h_gate, backend)
show_and_save_histogram(results, "h_gate_probabilities.png")
```

# SWAP

```python
swap_gate = QuantumCircuit(2)
swap_gate.x(0)
swap_gate.barrier()
swap_gate.swap(0, 1)
swap_gate.draw('mpl', filename=get_filename("swap_gate.png"))
```

```python
Statevector(swap_gate).draw('latex')
```

```python
transpiled_swap_gate = QuantumCircuit(2)
transpiled_swap_gate.x(0)
transpiled_swap_gate.barrier()
transpiled_swap_gate.cx(1,0)
transpiled_swap_gate.cx(0,1)
transpiled_swap_gate.cx(1,0)
transpiled_swap_gate.draw('mpl', filename=get_filename("transpiled_swap_gate.png"))
```

```python
Statevector(transpiled_swap_gate).draw('latex')
```
